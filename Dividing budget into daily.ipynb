```python
import numpy as np
import pandas as pd
sheet_id = "1zr67cSEIPI_H2MimMUoW_nNcFb0YHm3BVSz2_bTwE_4"
sheet_name = "Accounts"
url = f"https://docs.google.com/spreadsheets/d/{sheet_id}/gviz/tq?tqx=out:csv&sheet={sheet_name}"

df = pd.read_csv(url)

df[['EndDate','StartDate']] = df[['EndDate','StartDate']].apply(pd.to_datetime) #if conversion required
df['duration'] = (df['EndDate'] - df['StartDate']).dt.days

df['Final'] = df['Final'].astype(np.int64)
df['daily_budget'] = df['Final']/df['duration']

df.head()
```

```python
import datetime
final_data = []

for x in range(len(df['Marketerid'])):
    for i in range(df['duration'][x]):
#         print(df['Marketername'][x])
        data = {
            'Marketerid' : df['Marketerid'][x],
            'Marketername' : df['Marketername'][x],
            'Currency' : df['Currency'][x],
            'Direct/Agency' : df['Direct/Agency'][x],
            'Status' : df['Status'][x],
            'Account Manager' : df['Account Manager'][x],
            'Sales Manager' : df['Sales Manager'][x],
            'Budget' : df['Budget'][x],
            'Final' : df['Final'][x],
            'Upsale' : df['Upsale'][x],
            'Margin' : df['Margin'][x],
            'VD' : df['VD'][x],
            'Region' : df['Region'][x],
            'Category' : df['Category'][x],
            'StartDate' : df['StartDate'][x],
            'EndDate' : df['EndDate'][x],
            'duration' : df['duration'][x],
            'daily_budget' : df['daily_budget'][x],
            'Date' : df['StartDate'][x] + datetime.timedelta(days=i)
            
        }
        final_data.append(data)
df2 = pd.DataFrame.from_dict(final_data)
df2['StartDate'] = df2['StartDate'].dt.strftime('%Y%m%d%H%M%S')
df2['EndDate'] = df2['EndDate'].dt.strftime('%Y%m%d%H%M%S')
df2['Date'] = df2['Date'].dt.strftime('%Y%m%d%H%M%S')
df2.to_csv('C:/Users/Gajraj Singh/Downloads/ob-final_data.csv',index = False)
```
