```python
import os
import glob
import pandas as pd
import re
os.chdir('C:/Users/Gajraj Singh/Downloads')
extension = 'csv'

dfs = []
for i in glob.glob('*byDate_tyroo_*.{}'.format(extension)):
    j = re.findall("byDate_tyroo_(\w+)_2021", i)
    df = pd.read_csv(i,encoding = "ISO-8859-1")
    df['Account_Name'] = j[0]
    df.rename(columns = {df.columns[0]: 'id',}, inplace = True)
    df = df[df.id != 'total']
    df = df.iloc[: , 1:]
    df = df.fillna(0)
    dfs.append(df)
new_df = pd.concat(dfs, axis=0, ignore_index=True)
new_df = new_df[['Account_Name','From Date','To Date','Amount Spent','Impressions','Clicks','Conversions (click)'
                 ,'Conversions Value (click)','FirstDeposit Conversions (click)','FirstDeposit Conversions value (click)',
                 'OrderPlaced Conversions (click)','Addtocart Conversions (click)']]
new_df = new_df.fillna(0)
new_df.head()
```

```python
new_df.to_csv('C:/Users/Gajraj Singh/Downloads/new_df.csv', index=False,encoding = "ISO-8859-1")
```
