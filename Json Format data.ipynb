```python
import pandas as pd
import json
```

```python
df = pd.read_csv('C:/Users/Gajraj Singh/Documents/6eco_snap.csv')
```

```python
df.head()
```

```python
df['quantity'] = df['custom_data'].apply(lambda x:json.loads(x).get('quantity','NULL'))
df['size'] = df['custom_data'].apply(lambda x:json.loads(x).get('size','NULL'))
df['country'] = df['custom_data'].apply(lambda x:json.loads(x).get('country','NULL'))
df['category'] = df['custom_data'].apply(lambda x:json.loads(x).get('category','NULL'))
df['subcategory'] = df['custom_data'].apply(lambda x:json.loads(x).get('subcategory','NULL'))
df['language'] = df['custom_data'].apply(lambda x:json.loads(x).get('language','NULL'))
df['product_name'] = df['custom_data'].apply(lambda x:json.loads(x).get('product_name','NULL'))
df['brand_name'] = df['custom_data'].apply(lambda x:json.loads(x).get('brand_name','NULL'))
df['currency'] = df['custom_data'].apply(lambda x:json.loads(x).get('currency','NULL'))
df['product_sku'] = df['custom_data'].apply(lambda x:json.loads(x).get('product_sku','NULL'))
df['product_url'] = df['custom_data'].apply(lambda x:json.loads(x).get('product_url','NULL'))
```
