```python
from json import loads

def parser(data) :
    try :
        products = []
        i = 0
        data = loads(data.encode('utf-8'))
        cols = ['id', 'name', 'qty', 'brand', 'price'] 
        while True :
            key = 'product.{}.id'.format(i)
            if key in data :
                info = {}
                for col in cols :
                    info[col] = data.get('product.{}.{}'.format(i, col))
                products.append(info)
            else :
                break
            i += 1
        return products
    except :
        return []

df['clean'] = df['eventItemJson'].apply(lambda x:parser(x))
```

```python
data = df['clean'].values.tolist()
data
```

```python
Data = []
for i in data :
    for j in i :
        Data.append(j)
```

```python
res = pd.DataFrame(Data)
res.head()
```

```python
res['qty'].grouby()
```

```python
import numpy as np
out = pd.pivot_table(res, values=['qty','price'], index=['id','name'], aggfunc=np.sum, fill_value=0)
```
