```python
import pandas as pd
import json
```

```python
df = pd.read_csv('C:/Users/Gajraj Singh/Documents/6eco_snap_p.csv')
```

```python
df.head()
```

```python
df['final_amount'] = df['custom_data'].apply(lambda x:json.loads(x).get('final_amount','NULL'))
df['shipping_fee'] = df['custom_data'].apply(lambda x:json.loads(x).get('shipping_fee','NULL'))
df['discounted_amount'] = df['custom_data'].apply(lambda x:json.loads(x).get('discounted_amount','NULL'))
df['product_count'] = df['custom_data'].apply(lambda x:json.loads(x).get('product_count','NULL'))
df['currency'] = df['custom_data'].apply(lambda x:json.loads(x).get('currency','NULL'))
df['product_skus'] = df['custom_data'].apply(lambda x:json.loads(x).get('product_skus','NULL'))
df['total_amount'] = df['custom_data'].apply(lambda x:json.loads(x).get('total_amount','NULL'))
df['product_names'] = df['custom_data'].apply(lambda x:json.loads(x).get('product_names','NULL'))
df['value'] = df['custom_data'].apply(lambda x:json.loads(x).get('value','NULL'))
```
